```python
import glob
from collections import Counter
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Make expressed transcript table

```python
for path in sorted(glob.glob("../3_pseudobulk/results/pseudobulk/*.transcript_based.annotated.csv")):
    name = path.split("/")[-1].split(".transcript_based.")[0]
    d = pd.read_csv(path, index_col=0)
    d["Expressed"] = (d["TPM"] >= 10) & (d["Total"] >= 10)
    d.to_csv("results/expressed_transcripts/%s.csv" % name)
```

# Plot expressed transcripts per gene

```python
for path in sorted(glob.glob("results/expressed_transcripts/*.csv")):
    name = path.split("/")[-1][:-4]

    print("-" * 80)
    print("Name:", name)
    
    dat = pd.read_csv(path, index_col=0)
    dat = dat[dat["Expressed"]]
    data = dict()
    for gid, d in dat.groupby(by="GeneID"):
        data[gid] = d

    counter = Counter([len(d) for d in data.values()])
    n1 = counter[1]
    n2 = sum(counter.values()) - n1
    print("Genes expressed single isoform: %d" % n1)
    print("Genes expressed multiple isoforms: %d" % n2)

    xs = [1, 2, 3, 4, 5, 6]
    ys = [counter[x] for x in xs]
    xticks = [str(x) for x in xs]
    ys[-1] = sum(counter.values()) - sum(ys[:-1])
    xticks[-1] = "≥%s" % xticks[-1]

    plt.figure(figsize=(3, 3))
    plt.bar(xs, ys, edgecolor="black")
    plt.xlabel("Expressed isoforms per gene")
    plt.ylabel("Number of gene")
    plt.xticks(xs, xticks)
    plt.ylim(0, 6000)
    ys1 = np.arange(0, plt.gca().get_ylim()[1] + 1, 1000)
    plt.yticks(ys1, [format(int(y), ",") for y in ys1])
    plt.gca().spines["top"].set_visible(False)
    plt.gca().spines["right"].set_visible(False)
    plt.tight_layout()
    plt.savefig("figures/expressed_transcripts_per_gene/expressed_transcripts_per_gene.%s.pdf" % name)
    plt.show()
    plt.close()
```

# Analysis transcript type combination

```python
for path in sorted(glob.glob("results/expressed_transcripts/*.csv")):
    name = path.split("/")[-1][:-4]
    print("-" * 80)
    print("Name:", name)
    dat = pd.read_csv(path, index_col=0)
    dat = dat[dat["Expressed"]]
    data = dict()
    for gid, d in dat.groupby(by="GeneID"):
        data[gid] = d

    rows = []
    for gid, d in data.items():
        ss = []
        for k, v in sorted(Counter(d["TranscriptType"]).items()):
            ss.append("%s:%d" % (k, v))
        rows.append([gid, len(d), ";".join(ss)])
    d = pd.DataFrame(rows, columns=["GeneID", "TranscriptCount", "TranscriptTypes"])
    d.to_csv("results/expressed_transcript_type/%s.csv" % name, index=False)
    
    counter = Counter([tuple(list(sorted((d["TranscriptType"])))) for d in data.values()])
    items = list(sorted(counter.items(), key=lambda item: item[1], reverse=True))
    with open("results/expressed_transcript_type/%s.stats.csv" % name, "w+") as fw:
        fw.write("ID,GeneCount,TranscriptTypes\n")
        for i, (k, v) in enumerate(items):
            fw.write("%d,%d,%s\n" % (i, v, " + ".join(k)))
        for i, (k, v) in enumerate(items[:10]):
            print(i, v, k, sep="\t")
```

```python
# K562
data1 = [
    [3957, "protein_coding"],
    [575, "protein_coding × 2"],
    [164, "lncRNA"],
    [124, "protein_coding × 3"],
    [69, "protein_coding + processed_transcript"],
    [64, "protein_coding + nonsense_mediated_decay"],
    [56, "processed_transcript"],
    [54, "protein_coding + retained_intron"],
    [30, "protein_coding × 4"],
    [22, "protein_coding × 4 + nonsense_mediated_decay"]
]

# mESC
data2 = [
    [4850, "protein_coding"],
    [345, "protein_coding × 2"],
    [86, "processed_pseudogene"],
    [52, "protein_coding × 3"],
    [51, "protein_coding + retained_intron"],
    [35, "processed_transcript"],
    [34, "lincRNA"],
    [29, "protein_coding + processed_transcript"],
    [27, "protein_coding + nonsense_mediated_decay"],
    [17, "retained_intron"]
]
for cell_line in ["K562", "mESC"]:
    if cell_line == "K562":
        data = data1
        lim = 6000
    else:
        data = data2
        lim = 6000
    ys = np.arange(len(data))
    xs = [v[0] for v in data]
    ticks = [v[1] for v in data]
    
    plt.figure(figsize=(6.2, 3))
    plt.barh(ys, xs, edgecolor="black")
    for x, y in zip(xs, ys):
        plt.text(x + lim * 0.01, y, "%s" % format(x, ","), va="center")
    plt.xlim(0, lim)
    plt.ylim(max(ys) + 0.5, min(ys) - 0.5)
    plt.xlabel("Number of gene")
    plt.ylabel("Transcript type of expressed isoforms")
    plt.yticks(ys, ticks)
    xs1 = np.arange(0, lim + 1, 2000)
    plt.xticks(xs1, [format(int(x), ",") for x in xs1])
    plt.gca().spines["top"].set_visible(False)
    plt.gca().spines["right"].set_visible(False)
    plt.tight_layout()
    plt.savefig("figures/transcript_type_combination.%s.pdf" % cell_line)
    plt.show()
    plt.close()
```
