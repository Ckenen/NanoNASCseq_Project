```python
import anndata as ad
import scanpy as sc
import matplotlib.pyplot as plt
```

# 1. Gene-based

```python
adata = sc.read_h5ad("results/h5ad_clustered/Blastocyst.gene_based.h5ad")
```

```python
markers = [
    "Nanog", 
    "Sox2", 
    "Tdgf1", 
    "Gata4", 
    "Sox17", 
    "Cdx2", 
    "Gata3", 
    "Eomes", 
    "Elf5", 
    "Tfap2a",
    "BC051665", 
    "H19"
]
```

```python
with plt.rc_context({"figure.figsize": (2, 2)}):
    fig = sc.pl.umap(
        adata, 
        color=markers, 
        s=20, 
        frameon=False, 
        ncols=4, 
        vmax="p99", 
        save="_of_marker_genes.gene_based.pdf", 
        wspace=0.2, 
        hspace=0.3,
    )
```

```python
order = ["ICM-1", "ICM-2", "EPI", "PE", "TE-1", "TE-2", "muralTE", "polarTE"]
sc.pl.dotplot(
    adata, 
    markers, 
    "CellType", 
    categories_order=order, 
    swap_axes=False, 
    dendrogram=False, 
    figsize=(5, 2.5), 
    save="of_marker_genes.gene_based.pdf",
)
```
