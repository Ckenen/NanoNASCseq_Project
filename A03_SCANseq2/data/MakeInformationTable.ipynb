```python
import os
import pandas as pd
```

```python
infile = "Supplementary_Table_S1_Summary_of_sequenced_cells.xlsx"
outfile = "SCANseq2.xlsx"

if not os.path.exists(outfile):
    d = pd.read_excel(infile)
    d = d[d["Library"].isin(['4CL', '9CL', 'UMI_100', 'UMI_200'])]
    
    runs = []
    cells = []
    for lib, name, cell_line, species in d[["Library", "Name", "Cell_Line", "Organism"]].values:
        if lib == "UMI_200":
            lib = "SRR19279042_SCANseq2_UMI200"
        elif lib == "UMI_100":
            lib = "SRR19279044_SCANseq2_UMI100"
        elif lib == "4CL":
            lib = "SRR19279045_SCANseq2_4CL"
        elif lib == "9CL":
            lib = "SRR19279047_SCANseq2_9CL"
        
        cell = "%s.%s_%s_%s" % (lib, name, species, cell_line)
        runs.append(lib)
        cells.append(cell)
    d["Run"] = runs
    d["Cell"] = cells
    
    d1 = d[["Run", "Cell", "Organism", "Cell_Line", "Barcode.5", "Barcode.3"]]
    d1.to_excel(outfile, index=False)
```
