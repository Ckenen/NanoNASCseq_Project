```python
import numpy as np
import pandas as pd
```

# Cell lines

```python
info = pd.read_csv("../../1_NanoNASCseq/reports/NanoNASCseq_Summary.csv", index_col=0)
```

```python
# K562

for (s4u, time), d in info[info["CellType"] == "K562"].groupby(by=["s4U", "Time"]):
    d = d[(d["UMIs"] > 5000) & (d["ActD"].isna())]
    if time >= 1:
        d.to_csv("results/tables/K562.s4U_%duM_%dh.csv" % (s4u, time))
        print("K562 (%duM, %dh):" % (s4u, time), d.shape)
    else:
        d.to_csv("results/tables/K562.s4U_%duM_%dmin.csv" % (s4u, time*60))
        print("K562 (%duM, %dmin):" % (s4u, time*60), d.shape)
    
d = info
d = d[(d["CellType"] == "K562") & (d["s4U"].isin([0, 50])) & (d["Time"] == 3) & (d["UMIs"] > 5000) & (d["ActD"].isna())]
d.to_csv("results/tables/K562.csv")
print("K562:", d.shape)

d = info
d = d[(d["CellType"] == "K562") & (d["s4U"] == 50) & (d["Time"] == 3) & (d["UMIs"] > 5000) & (d["ActD"].isna()) & (d["TC.Ratio"] > 0.008)]
d.to_csv("results/tables/K562.s4U_50uM_3h.highTC.csv")
print("K562 (treatment, TC%>0.8%):", d.shape)
```

```python
# mESC

for (s4u, time), d in info[info["CellType"] == "mESC"].groupby(by=["s4U", "Time"]):
    d = d[(d["UMIs"] > 5000) & (d["ActD"].isna())]
    d.to_csv("results/tables/mESC.s4U_%duM_%dh.csv" % (s4u, time))
    print("mESC (%duM, %dh):" % (s4u, time), d.shape)

d = info
d = d[(d["CellType"] == "mESC") & (d["s4U"].isin([0, 400])) & (d["Time"] == 3) & (d["UMIs"] > 5000) & (d["ActD"].isna())]
d.to_csv("results/tables/mESC.csv")
print("mESC (0uM and 400uM, 3h):", d.shape)

d = info
d = d[(d["CellType"] == "mESC") & (d["s4U"] == 400) & (d["Time"] == 3) & (d["UMIs"] > 5000) & (d["ActD"].isna()) & (d["TC.Ratio"] > 0.01)]
d.to_csv("results/tables/mESC.s4U_400uM_3h.highTC.csv")
print("mESC (400uM, 3h, TC%>1%):", d.shape)
```

```python
# K562-FUCCI

for (s4u, time, phase), d in info[info["CellType"] == "K562_FUCCI"].groupby(by=["s4U", "Time", "Phase"]):
    d = d[(d["UMIs"] > 5000) & (d["ActD"].isna())]
    d.to_csv("results/tables/K562_FUCCI_%s.s4U_%duM_%dh.csv" % (phase, s4u, time))
    print("K562-FUCCI (%s, %suM, %dh):" % (phase, s4u, time), d.shape)
```

```python
for (s4u, time, iaa), d in info[info["CellType"] == "mESC_EXOSC2"].groupby(by=["s4U", "Time", "IAA"]):
    d = d[(d["UMIs"] > 5000) & (d["ActD"].isna())]
    d.to_csv("results/tables/mESC_EXOSC2_IAA_%dh.s4U_%duM_%dh.csv" % (iaa, s4u, time))
    print("mESC_EXOSC2 (IAA %dh, %suM, %dh):" % (iaa, s4u, time), d.shape)
```

# Mouse blastocyst

```python
# Blastocyst

info = pd.read_csv("../2_blastocyst/1_clustering/results/h5ad_consensus/Blastocyst.gene_based.obs.csv", index_col=0)
for (s4u, time, ct), d in info.groupby(by=["s4U", "Time", "CellType"]):
    if s4u != 400 or ct == "Conflict":
        continue
    d = d[(d["UMIs"] > 5000) & (d["ActD"].isna()) & (d["TC.Ratio"] >= 0.002)]
    d.to_csv("results/tables/MouseBlastocyst_%s.s4U_%duM_%dh.csv" % (ct, s4u, time))
    print("%s (%duM, %dh):" % (ct, s4u, time), d.shape)
```

```python
tmp = info[(info["s4U"] == 400) & (info["Time"] == 3) & (info["ActD"].isna()) & (info["TC.Ratio"] >= 0.002)]

d = tmp[tmp["CellType"].isin(["ICM-1", "ICM-2"])]
print("ICM", d.shape, sep="\t")
d.to_csv("results/tables/MouseBlastocyst_ICM.s4U_400uM_3h.csv")

d = tmp[tmp["CellType"].isin(["TE-1", "TE-2"])]
print("EarlyTE", d.shape, sep="\t")
d.to_csv("results/tables/MouseBlastocyst_EarlyTE.s4U_400uM_3h.csv")

d = tmp[tmp["CellType"].isin(["polarTE", "muralTE"])]
print("LateTE", d.shape, sep="\t")
d.to_csv("results/tables/MouseBlastocyst_LateTE.s4U_400uM_3h.csv")
```
