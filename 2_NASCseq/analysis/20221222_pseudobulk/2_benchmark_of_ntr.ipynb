```python
import pandas as pd
```

```python
params = [
    ["ENST00000652288.1", "MYC"],
    ["ENST00000317968.9", "PDLIM5"],
    ["ENST00000646664.1", "ACTB"],
    ["ENST00000229239.10", "GAPDH"]
]
for name in ["GSE128273_NASCseq.s4U_50uM_3h", "NASCseq.s4U_50uM_3h"]:
    print("Name:", name)
    path = "results/pseudobulk/%s.tsv.gz" % name
    d = pd.read_csv(path, sep="\t")
    print("Gene\tTotal\tNew\tNTR")
    print("-" * 80)
    for tid, gname in params:
        d1 = d[d["TranscriptID"] == tid]
        v1 = d1["Count"].values[0]
        v2 = d1["Count.New"].values[0]
        print(gname, v1, v2, v2/v1, sep="\t")
    print("-" * 80)
```
