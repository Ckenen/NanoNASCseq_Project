```python
from collections import Counter, defaultdict
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
```

# 1. UMAP of cell types (gene-based)

```python
adata = sc.read_h5ad("results/h5ad_clustered/Blastocyst.gene_based.h5ad")
```

```python
with plt.rc_context({"figure.figsize": (4, 4)}):
    sc.pl.umap(
        adata, 
        color="CellType", 
        frameon=True, 
        size=50, 
        title="Cell type", 
        legend_loc="on data", 
        legend_fontoutline=2, 
        save="_of_cell_type.gene_based.pdf",
    )
```
